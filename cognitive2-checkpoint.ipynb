```python
ac=[]

import random
from sklearn.linear_model import Ridge
model = Ridge(alpha=1.0)

def objective_function(x):
    return x**2  


class Particle:
    def __init__(self):
        self.position = random.uniform(-5, 5)  
        self.velocity = random.uniform(-1, 1)  
        self.best_position = self.position  
        self.best_score = float('inf')  #

    def particle_swarm_optimization(self,num_particles, num_iterations):
        particles = [Particle() for _ in range(num_particles)]  
        global_best_position = None  
        global_best_score = float('inf')  

        for _ in range(num_iterations):
            for particle in particles:
                score = objective_function(particle.position)  
                if score < particle.best_score:
                    particle.best_score = score
                    particle.best_position = particle.position

                if score < global_best_score:
                    global_best_score = score
                    global_best_position = particle.position


                inertia_weight = 0.5
                cognitive_weight = 0.5
                social_weight = 0.5
                particle.velocity = (inertia_weight * particle.velocity +
                                     cognitive_weight * random.random() * (particle.best_position - particle.position) +
                                     social_weight * random.random() * (global_best_position - particle.position))
                particle.position += particle.velocity

        return global_best_position, global_best_score

    def train(self,X_train,y_train,X_test):
        model.fit(X_train,y_train)
        y_pred=model.predict(X_test)
        return y_pred

num_particles = 30  
num_iterations = 100 
p=Particle()
best_position, best_score =p.particle_swarm_optimization(num_particles, num_iterations)

y_pred=p.train(X_train,y_train,X_test)
print("Best Position:", best_position)
print("Best Score:", best_score)
ac.append(accuracy_score(y_pred,y_train,sample_weight=0.2)*100)
```

```python
import random

class AntColony:
    def __init__(self, distances, num_ants=10, evaporation_rate=0.5, pheromone_constant=1.0, heuristic_constant=1.0):
        self.distances = distances
        self.num_ants = num_ants
        self.evaporation_rate = evaporation_rate
        self.pheromone_constant = pheromone_constant
        self.heuristic_constant = heuristic_constant
        self.pheromone_levels = [[1.0 for _ in range(len(distances[0]))] for _ in range(len(distances))]

    def _calculate_probability(self, current_city, unvisited_cities):
        probabilities = []
        for city in unvisited_cities:
            pheromone = self.pheromone_levels[current_city][city]
            heuristic = 1 / self.distances[current_city][city]
            probability = (pheromone ** self.pheromone_constant) * (heuristic ** self.heuristic_constant)
            probabilities.append((city, probability))
        total_probability = sum(prob for _, prob in probabilities)
        probabilities = [(city, prob / total_probability) for city, prob in probabilities]
        return probabilities

    def run(self, num_iterations):
        for _ in range(num_iterations):
            solutions = []
            for _ in range(self.num_ants):
                current_city = random.randint(0, len(self.distances[0]) - 1)
                unvisited_cities = set(range(len(self.distances[0]))) - {current_city}
                solution = [current_city]

                while unvisited_cities:
                    probabilities = self._calculate_probability(current_city, unvisited_cities)
                    next_city = random.choices([city for city, _ in probabilities], [prob for _, prob in probabilities])[0]
                    solution.append(next_city)
                    unvisited_cities.remove(next_city)
                    current_city = next_city

                solutions.append(solution)

            for i in range(len(self.distances)):
                for j in range(len(self.distances[i])):
                    self.pheromone_levels[i][j] *= (1 - self.evaporation_rate)
                    for solution in solutions:
                        if (i, j) in zip(solution, solution[1:] + [solution[0]]):
                            self.pheromone_levels[i][j] += (self.pheromone_constant / self.distances[i][j])

        best_solution = min(solutions, key=lambda x: sum(self.distances[i][j] for i, j in zip(x, x[1:] + [x[0]])))
        best_distance = sum(self.distances[i][j] for i, j in zip(best_solution, best_solution[1:] + [best_solution[0]]))
        return best_solution, best_distance
    def train(self,X_train,y_train,X_test):
        model.fit(X_train,y_train)
        y_pred=model.predict(X_test)
        return y_pred


    
distances = [[0, 10, 15, 20],[10, 0, 35, 25],[15, 35, 0, 30],[20, 25, 30, 0]]

num_ants = 10
num_iterations = 100

ant_colony = AntColony(distances, num_ants=num_ants)
best_solution, best_distance = ant_colony.run(num_iterations)
p=AntColony(distances, num_ants=num_ants)
y_pred=p.train(X_train,y_train,X_test)
print("Best TSP Solution:", best_solution)
print("Total Distance:", best_distance)
ac.append(accuracy_score(y_pred,y_train,sample_weight=0.8)*100)
```

```python
from scipy import signal
import numpy as np
sig = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)
sig_noise = sig + np.random.randn(len(sig))
corr = signal.correlate(sig_noise, np.ones(128), mode='same') / 128
import matplotlib.pyplot as plt
clock = np.arange(64, len(sig), 128)
fig, (ax_orig, ax_noise, ax_corr) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.plot(clock, sig[clock], 'ro')
ax_orig.set_title('Original signal')
ax_noise.plot(sig_noise)
ax_noise.set_title('Signal with noise')
ax_corr.plot(corr)
ax_corr.plot(clock, corr[clock], 'ro')
ax_corr.axhline(0.5, ls=':')
ax_corr.set_title('Cross-correlated with rectangular pulse')
ax_orig.margins(0, 0.1)
fig.tight_layout()
fig.show()
```
